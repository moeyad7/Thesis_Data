## Pre-trained Models

### Scatter Plot

```python
import matplotlib.pyplot as plt

ANERCorp_F1 = [0.158, 0.098, 0.21, 0.294]
ArSATwitter_F1 = [0.539, 0.438, 0.528, 0.728]
ArEnSA_F1 = [0.333, 0.412, 0.365, 0.504]

models = ['roberta-base', 'albert-base-v2', 'arabert-base', 'arabert-twitter-base']

plt.figure(figsize=(6, 3))

plt.scatter(models, ANERCorp_F1, label='ANERCorp')
plt.scatter(models, ArSATwitter_F1, label='ArSATwitter')
plt.scatter(models, ArEnSA_F1, label='ArEnSA')

plt.xlabel('Models')
plt.ylabel('F1-Score')
plt.title('F1-Score for Different Models')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

### Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

ANERCorp_F1 = [0.158, 0.098, 0.21, 0.294]
ArSATwitter_F1 = [0.539, 0.438, 0.528, 0.728]
ArEnSA_F1 = [0.333, 0.412, 0.365, 0.504]

models = ['roberta-base', 'albert-base-v2', 'arabert-base', 'arabert-twitter-base']

bar_width = 0.2  # Adjust the bar width if needed
index = np.arange(len(models))

plt.figure(figsize=(10, 5))

# Use plt.bar with adjusted positions
plt.bar(index - bar_width, ANERCorp_F1, width=bar_width, label='ANERCorp', alpha=0.7)
plt.bar(index, ArSATwitter_F1, width=bar_width, label='ArSATwitter', alpha=0.7)
plt.bar(index + bar_width, ArEnSA_F1, width=bar_width, label='ArEnSA', alpha=0.7)

plt.xlabel('Models')
plt.ylabel('F1-Score')
plt.title('F1-Score for Different Models')
plt.xticks(index, models)
plt.legend()
plt.grid(axis='y')
plt.show()
```

## Learning Rate

```python
import matplotlib.pyplot as plt

ANERCorp_F1 = [0.294, 0.326, 0.323]
ArSATwitter_F1 = [0.728, 0.721, 0.717]
ArEnSA_F1 = [0.504, 0.512, 0.584]

learning_rate = [1e-5, 2e-5, 3e-5]

plt.figure(figsize=(6, 3))

plt.plot(learning_rate, ANERCorp_F1, marker='o', label='ANERCorp')
plt.plot(learning_rate, ArSATwitter_F1, marker='o', label='ArSATwitter')
plt.plot(learning_rate, ArEnSA_F1, marker='o', label='ArEnSA')

plt.xlabel('Learning Rate')
plt.ylabel('F1-Score')
plt.title('F1-Score for Different Learning Rates')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

## Gradient Accumulation Steps

```python
import matplotlib.pyplot as plt

ANERCorp_F1 = [0.326, 0.295, 0.182]
ArSATwitter_F1 = [0.728, 0.735, 0.719]
ArEnSA_F1 = [0.584, 0.441, 0.364]

gradient_accumulation_steps = [1, 2, 4]

plt.figure(figsize=(6, 3))

plt.plot(gradient_accumulation_steps, ANERCorp_F1, marker='o', label='ANERCorp')
plt.plot(gradient_accumulation_steps, ArSATwitter_F1, marker='o', label='ArSATwitter')
plt.plot(gradient_accumulation_steps, ArEnSA_F1, marker='o', label='ArEnSA')

plt.xlabel('Gradient Accumulation Steps')
plt.ylabel('F1-Score')
plt.title('F1-Score for Different Gradient Accumulation Steps')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(gradient_accumulation_steps)  # Set x-axis ticks explicitly
plt.grid(True)
plt.show()
```
