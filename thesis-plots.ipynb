## Learning Rate

```python
import matplotlib.pyplot as plt

ANERCorp_F1 = [0.294, 0.326, 0.323]
ArSATwitter_F1 = [0.728, 0.721, 0.717]
ArEnSA_F1 = [0.504, 0.512, 0.584]
ArEnNER_F1 = [0.270,0.344,0.333]

learning_rate = ["1E-5", "2E-5", "5E-5"]

plt.figure(figsize=(6, 3))  # Increase figure size for better readability

# Use a loop for cleaner code
for data, label,color in zip([ANERCorp_F1, ArSATwitter_F1, ArEnSA_F1,ArEnNER_F1], ['ANERCorp', 'ArSATwitter', 'ArEnSA','ArEnNER'],["royalblue","#FF7F0E","green","purple"]):
    plt.plot(learning_rate, data, marker='o', label=label,color=color)

plt.xlabel('Learning Rate')
plt.ylabel('F1-Score')
plt.title('F1-Score for Different Learning Rates')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(learning_rate, rotation=45, ha="right")

# Add grid lines to the y-axis for better readability
plt.gca().yaxis.grid(True, linestyle='--', alpha=0.6)

# Add a horizontal line for reference (e.g., F1-Score of 0.5)
plt.axhline(y=0.5, color='gray', linestyle='--', linewidth=1, alpha=0.7)

plt.tight_layout()  # Adjust layout to prevent clipping labels
plt.show()
```
